### Importing libraries and datasets

```python
import pandas as pd
```

```python
#data of match statistics

matchstatsl = pd.read_csv("matchstatsl.csv")

playerstatsl = pd.read_csv("slbatstatmat.csv")
```

### Data cleaning

#### Cleaning matchstatsl dataset

```python
# removing unimportant columns

matchstatsl = matchstatsl.drop(["web-scraper-order","web-scraper-start-url"], axis=1)
matchstatsl
```

```python
# Remove 'DNB' 

matchstatsl = matchstatsl[matchstatsl['Score'] != 'DNB']
```

```python
matchstatsl.tail()
```

```python
# make Score, RPO numberical.

matchstatsl['Scorecopy'] = matchstatsl['Score'].str.split('/').str[0]
matchstatsl.head()
```

```python
# making a new columns for wickets fallen and filling nan with 10 wickets

matchstatsl['Wicketslost'] = matchstatsl['Score'].str.split('/').str[1]
matchstatsl['Wicketslost'] = matchstatsl['Wicketslost'].fillna(10)
matchstatsl
```

```python
#changing type of new columns

matchstatsl['Wicketslost'] = matchstatsl['Wicketslost'].astype('int')
matchstatsl['Scorecopy'] = matchstatsl['Scorecopy'].astype('int')
```

```python
# removine v form opposition column

matchstatsl['OppositionCopy'] = matchstatsl['Opposition'].str.strip(' ').str[1:]
matchstatsl.head()
```

```python
matchstatsl['Dateinymdformat'] = pd.to_datetime(matchstatsl['Start Date'])
matchstatsl
```

```python
# removing unwanted columns

matchstatsl_refined = matchstatsl.loc[:, ['Dateinymdformat','OppositionCopy', 'Ground' ,'Inns' ,'Scorecopy', 'Wicketslost', 'RPO', 'Overs', 'Result']]
```

```python
matchstatsl_refined
```

```python
matchstatsl['Start Date'].nunique()
```

#### cleaning palyerstatsl dataset

```python
# removing unimportant columns

playerstatsl = playerstatsl.drop(["web-scraper-order","web-scraper-start-url"], axis=1)
playerstatsl
```

```python
playerstatsl['Date'].nunique()
```

```python
# remove team did not bat data

playerstatsl = playerstatsl[playerstatsl['Runs'] != 'TDNB']

playerstatsl['Date'].nunique()

# it seems some players have scored in abandoned matches also because unique count of cleared match stat is 223 
```

```python
# making RunsCopy and replace 'DNB' with 0 and remove (*)

playerstatsl['Runscopy'] = playerstatsl['Runs']
playerstatsl['Runscopy'] = playerstatsl['Runscopy'].replace(['DNB'],'0')
playerstatsl['Runscopy'] = playerstatsl['Runscopy'].str.split('*').str[0]
playerstatsl['Runscopy'] = playerstatsl['Runscopy'].astype('int')
playerstatsl.info()
```

```python
playerstatsl['Runs'].unique()
```

```python
# Add a column not out and if a star is found in Runs put not out in the field

playerstatsl['Not Out'] = playerstatsl['Runs'].str.contains('\*') 
playerstatsl['Batted'] = ~playerstatsl['Runs'].str.contains('DNB') 
# or playerstatsl['Runs'].str.contains('DNB')

playerstatsl[0:50]
    
    
```

```python
playerstatsl[0:50]
```

```python
# make copies and fill empty entries of Balls Faced and Strike Rate and make the int or float

playerstatsl['Balls Faced Copy'] = playerstatsl['Balls Faced'].fillna(10)
playerstatsl['Balls Faced Copy'] =playerstatsl['Balls Faced Copy'].replace(['-'],0)
playerstatsl['Strick Rate Copy'] = playerstatsl['Strick Rate'].fillna(10)
playerstatsl['Strick Rate Copy'] = playerstatsl['Strick Rate Copy'].replace(['-'],0)


playerstatsl['Balls Faced Copy'] = playerstatsl['Balls Faced Copy'].astype('int')
playerstatsl['Strick Rate Copy'] = playerstatsl['Strick Rate Copy'].astype('float')

playerstatsl.info()

```

```python
# removing v form opposition column

playerstatsl['OppositionCopy'] = playerstatsl['Opposition'].str.strip(' ').str[1:]
playerstatsl
```

```python
# Date string to a dateformat

playerstatsl['Dateinymdformat'] = pd.to_datetime(playerstatsl['Date'])
playerstatsl
```

```python
# cleaned dataframe preview

playerstatsl.head()
```

```python
# removing unwanted columns

playerstatsl_refined = playerstatsl.loc[:, ['Dateinymdformat','Player', 'OppositionCopy', 'Ground' ,'Inning' , 'Batted','Runscopy', 'Balls Faced Copy', 'Strick Rate Copy', 'Not Out']]
playerstatsl_refined.head()
```

### Merge dataframes

```python
# merge function will merge two dataframes according to dateinymdformat,renaming columns and drop duplicates

playermatchstatsl = pd.merge(playerstatsl_refined, matchstatsl_refined)
pd.set_option('display.max_columns', None)
playermatchstatsl.columns = ('Date', 'Player', 'Opposition', 'Ground', 'Inning',
       'Batted', 'Runs', 'Balls Faced', 'Strick Rate', 'Not Out',
       'Inns', 'TeamScore', 'TeamWicketslFallen', 'RPO', 'Overs', 'Result')
playermatchstatsl = playermatchstatsl.drop(['Inns'], axis = 1)
playermatchstatsl
```
